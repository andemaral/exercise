```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df=pd.read_csv('Interview Exercise Nick2.csv',parse_dates=['Start Date','End Date'])
df.dtypes
```

```python
df.shape
```

```python
df['Id']=df['Id'].astype('object')
df.dtypes
df['Duration']=df['End Date']-df['Start Date']
```

```python
df['Duration Years']=df['Duration']/pd.to_timedelta(1,unit="Y")
df['Duration Months']=df['Duration']//pd.to_timedelta(1,unit="M")
df.head(5)
print(df.shape)
```

```python
df.head(5)
```

```python
print(min(df['Start Date']))
print(max(df['End Date']))
```

```python
from datetime import timedelta 
def check(start,end):  
    Result = []
    while start <= end:
        Result.append(start)
        start = start + timedelta(days=1)
    return Result
```

```python
df['ActivityDate'] = df.apply(lambda row: check(row['Start Date'],row['End Date']),axis=1)
df = df.explode('ActivityDate')  
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.sort_values(by='ActivityDate',ascending=True)
df['ActivityDateDif']=df['ActivityDate']-df['Start Date']
df['ActivityDateDifMonth']=df['ActivityDateDif']/pd.to_timedelta(1,unit="M")
df['ActivityDateDifYear']=df['ActivityDateDif']/pd.to_timedelta(1,unit="Y")
#df['Job Ended']=df['ActivityDateDifMonth']>=df['Duration Months']
```

```python
df['day'] = df['ActivityDate'].apply(lambda x: x.day)
```

```python
df=df[df['day']==1]
```

```python
df['Job Ended']=df['End Date']<=df['ActivityDate']
```

```python
df.head(72)
```

```python
df['completed']=df['ActivityDateDif']*df['Job Ended']
df['completed']=df['completed']/pd.to_timedelta(1,unit="M")
```

```python
df.head(72)
```

```python
# making list of values   
# making series from list 
series =df['Job Ended']
#print(series)  
# calling method 
cumsum = round(series.cumsum(),4) 
df['cumsum']=cumsum
series_completed=df['completed']
cumsum2=round(series_completed.cumsum(),4)
df['cumsum_completed']=cumsum2
```

```python
df['avg_completed_duration']=df['cumsum_completed']/df['cumsum']
```

```python
#df.replace(0,np.nan)[df['Id']==1].groupby(['ActivityDate',])['completed'].agg({'completed': lambda x: x.mean(skipna=True)}).sort_values(by='ActivityDate',ascending=True).head(50)
#df.groupby['Id','ActivityDate'][]
```

```python
df['avg_completed_duration_year']=df['avg_completed_duration']/12
```

```python
df.sort_values(by=['ActivityDate','End Date'])
report=df[['ActivityDate','Start Date', 'End Date', 'Id',
       'Duration Months','ActivityDateDifYear','completed','avg_completed_duration_year']]
report.rename(columns={'ActivityDateDifYear':'Cumul Year Experience','avg_completed_duration_year':'Avg Completed Duration'},inplace=True)
```
